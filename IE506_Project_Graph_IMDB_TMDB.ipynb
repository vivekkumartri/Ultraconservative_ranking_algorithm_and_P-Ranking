```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
loss11=np.loadtxt('/home/22n0457/imdb/loss1.csv',delimiter=",")
loss21=np.loadtxt('/home/22n0457/imdb/loss2.csv',delimiter=",")
loss31=np.loadtxt('/home/22n0457/imdb/loss3.csv',delimiter=",")
loss41=np.loadtxt('/home/22n0457/imdb/loss4.csv',delimiter=",")
loss51=np.loadtxt('/home/22n0457/imdb/loss5.csv',delimiter=",")
loss61=np.loadtxt('/home/22n0457/imdb/loss6.csv',delimiter=",")
loss71=np.loadtxt('/home/22n0457/imdb/loss7.csv',delimiter=",")
loss81=np.loadtxt('/home/22n0457/imdb/loss8.csv',delimiter=",")
```

```python
plt.figure(figsize=(17,8))
plt.plot(loss11[::50],label='ranking loss of widrowhoff')
plt.plot(loss21[::50],label='ranking loss of Pranking')
plt.plot(loss31[::50],label='ranking loss of uniform additive multiclass algorithm')
plt.plot(loss41[::50],label='ranking loss of worst additive multiclass algorithm')
plt.plot(loss51[::50],label='ranking loss of similarity score additive multiclass algorithm')
plt.plot(loss61[::50],label='ranking loss of margin infused Relaxed Algorithm (MIRA)')
plt.plot(loss71[::50],label='ranking loss of Pranking updated')
plt.plot(loss81[:20],label='ranking loss neural ranking approach')
plt.title("code implementation on imdb Data set")
plt.xlabel("number of epochs")
plt.xticks(range(0,20))
plt.ylabel("Ranking loss")
plt.legend(bbox_to_anchor=(0.5,1))
plt.show()
```
