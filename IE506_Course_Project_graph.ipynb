```python
loss11=np.loadtxt('/home/22n0457/imdb/loss2.csv',delimiter=",")
```

```python
loss11=np.reshape(loss11,(20,50)).sort(axis=1)
```

```python
def ci(x):
    x=np.array(x)
    halfwidth=[]
    for i in range(x.shape[1]):
        a,b=st.t.interval(alpha=0.95, df=len(x[:,i])-1, loc=np.mean(x[:,i]), scale=st.sem(x[:,i]))
        halfwidth.append(np.mean(x[:,i])-a)
    return halfwidth
```

```python
plt.errorbar(np.arange(50000)[::1000],np.mean(loss11,axis=0),yerr=ci(loss11),label='time average ranking loss of wirowhoff', uplims=True, lolims=True)
plt.xlabel('Rounds')
plt.ylabel('Ranking loss')
plt.legend()
plt.show()
```
